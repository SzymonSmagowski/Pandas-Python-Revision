## Add and delete columns

```python
frame = pd.read_csv("./mcdonalds.csv",usecols=["Item","Category","Calories","TotalFat","Sugars","Protein","Cholesterol"])
```

```python
frame.head()
```

```python
frame.TotalFat.head()
```

```python
frame["TotalFat"].head()
```

```python
frame["NewCol"] = 0
```

```python
frame.head()
```

```python
frame["SugarsAndFat"] = frame.Sugars + frame.TotalFat
```

```python
frame.head()
```

```python
frame.insert(loc=2,column="SPF", value=frame.Sugars+frame.Protein+frame.TotalFat)
```

```python
frame.head()
```

```python
del frame["SugarsAndFat"]
```

```python
frame.head()
```

```python
frame = frame.drop(columns=["NewCol"])
```

```python
frame.head()
```

# Cleaning data

## Dropna()

```python
frame = pd.read_csv("./mcdonalds.csv")
```

```python
frame
```

```python
frame.iloc[1,1] = np.NaN
```

```python
frame.dropna() # deleting whole rowswith NaN
```

```python
frame.dropna(how='all').head()
```

```python
frame.dropna(subset=["Category"]).head()
```

```python
frame.dropna(axis='columns').head() # deleting column with NaN - axis parameter means rows or columns
```

## Fillna()

```python
frame.head()
```

```python
frame.fillna(value=0).head()
```
