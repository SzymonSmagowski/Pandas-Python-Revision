```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
frame = pd.read_csv("./WA_Sales_Products_2012-14.csv")
```

### Basically creating data frame from pivot table

```python
pt = frame.pivot_table(index='Retailer country', columns='Year', values='Revenue', aggfunc='sum')
pt.head()
```

```python
pt.reset_index(inplace=True)
pt.head()
```

```python
pt.fillna(0,inplace=True)
pt.head()
```

```python
pt.melt(id_vars='Retailer country',value_name='RevenueSum',var_name='YearOfTransaction', value_vars=[2012,2013,2014]).head()
```
