```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
frame = pd.read_csv("./WA_Sales_Products_2012-14.csv")
frame.head()
```

```python
groups = frame.groupby(by='Retailer country')
```

```python
for country in groups:
    print(country[0])
```

```python
country_names = []
for country in groups:
    country_names.append(country[0])
```

```python
for country in groups:
    print(country[1])
```

```python
for country in groups:
    print(country[0], len(country[1]))
```

```python
for country, country_data in groups:
    print(country, len(country_data), len(country)) # country_data - smaller df object having values for every country
```

```python
for country, country_data in groups:
    print(country,
         country_data['Revenue'].max() - country_data['Revenue'].min())
```

```python
for country, country_data in groups:
    print(country,
         country_data['Revenue'].max(),
         country_data['Revenue'].idxmax())
```

```python
for country,country_data in groups:
    print(country,
         country_data['Revenue'].max(),
         country_data['Revenue'].idxmax(),
         country_data.loc[country_data['Revenue'].idxmax()])
```

```python
the_biggest = pd.DataFrame()
for country,country_data in groups:
    the_biggest = the_biggest.append(country_data.nlargest(1,'Revenue'))
```
