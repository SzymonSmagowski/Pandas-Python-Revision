```python
groups['Revenue'].sum().head()
```

```python
groups.agg({"Revenue":"sum"}).head() # The same way of doing aggregate functions
```

```python
groups.agg({"Revenue":"sum",
           "Quantity":"min",
           "Gross margin":"max"}).head() # different functions for different columns
```

```python
groups.agg({"Revenue":["sum","min","mean"],
           "Quantity":"min",
           "Gross margin":"max"}).head()
```

```python
my_aggregation = ["min","max","mean"]
```
